```python
from docling.document_converter import DocumentConverter
from collections import defaultdict
from IPython.display import Markdown, display
import boto3
import dotenv
import loguru

dotenv.load_dotenv("../.env")

BUCKET_NAME = "llm-zoomcamp-pj2025"

class DocumentAnalyzer:

    def __init__(self, document_path: str, ) -> None:

        converter = DocumentConverter()
        result = converter.convert(document_path)
        loguru.logger.info("Processing Document...")
        self.converted_document = result.document

    def persist_in_s3(self, s3_path: str) -> None:
        
        s3_client = boto3.client('s3', region_name="us-east-1")
        s3_client.put_object(
            Bucket=BUCKET_NAME,
            Key=f"processed_documents/text_processed.md",
            Body=self.converted_document.export_to_markdown().encode('utf-8'),
            ContentType='text/markdown'
        )
```

```python
analyzer = DocumentAnalyzer(document_path="../data/Results_Report_1Q25.pdf")
```

```python
analyzer.persist_in_s3(s3_path="processed_documents/")
```
